```python
import pandas as pd

df_base = pd.read_hdf(data_path)
```

```python
df = df_base.copy()
```

```python
isotopes = ["Pu236", "Pu237", "Pu238", "Pu239", "Pu240", "Pu241", "Pu242", "Pu243", "Pu244", "Pu246"]

# Create patterns to keep
patterns_to_keep = [f"fuel_{isotope}" for isotope in isotopes] + [f"rank_{isotope}" for isotope in isotopes]

# Get columns that start with 'fuel_' or 'rank_'
fuel_tank_columns = [col for col in df.columns if col.startswith(('fuel_', 'tank_'))]

# Filter columns to drop (those that start with fuel_ or rank_ but are not in patterns_to_keep)
columns_to_drop = [col for col in fuel_tank_columns if col not in patterns_to_keep]

# Drop the filtered columns
df = df.drop(columns=columns_to_drop)
```

```python
# Get the remaining fuel_ and rank_ columns
remaining_fuel_tank_cols = [col for col in df.columns if col.startswith(('fuel_', 'tank_'))]

# Drop rows where all remaining fuel_ and rank_ columns are 0
df = df[~(df[remaining_fuel_tank_cols] == 0).all(axis=1)]

# Verify the results
print("Shape after cleaning:", df.shape)
print("\nRemaining fuel_ and tank_ columns:", remaining_fuel_tank_cols)
```
