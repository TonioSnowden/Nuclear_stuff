```python
# Convert particle_times to strings for comparison and count duplicates
duplicate_counts = data['particle_times'].astype(str).value_counts()

# Show how many times each unique particle_times sequence appears
print("Number of duplicate particle_times sequences:")
print(duplicate_counts)

# Show total number of unique sequences
print("\nTotal number of unique sequences:", len(duplicate_counts))
```
